```python
#KDD
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
# ==========================================
# STEP 1: DATA SELECTION & COLLECTION
# ==========================================
input_filename = "Top Indian Places to Visit.csv"
df = pd.read_csv(input_filename)
print("--- Step 1: Dataset Loaded Successfully ---")
print(df.head())
# ==========================================
# STEP 2: DATA CLEANING
# ==========================================
print("\n--- Step 2: Data Cleaning ---")
df_clean = df.dropna().copy()
print("Rows after cleaning:", len(df_clean))
# ==========================================
# STEP 3: DATA TRANSFORMATION
# ==========================================
print("\n--- Step 3: Data Transformation ---")
# Label Encoders
le_zone = LabelEncoder()
le_state = LabelEncoder()
le_city = LabelEncoder()
le_type = LabelEncoder()
le_airport = LabelEncoder()
le_weeklyoff = LabelEncoder()
le_significance = LabelEncoder()
le_dslr = LabelEncoder()
le_besttime = LabelEncoder()
# Encode categorical columns
df_clean["Zone_Encoded"] = le_zone.fit_transform(df_clean["Zone"])
df_clean["State_Encoded"] = le_state.fit_transform(df_clean["State"])
df_clean["City_Encoded"] = le_city.fit_transform(df_clean["City"])
df_clean["Type_Encoded"] = le_type.fit_transform(df_clean["Type"])
df_clean["Airport_Encoded"] = le_airport.fit_transform(df_clean["Airport with 50km Radius"])
df_clean["WeeklyOff_Encoded"] = le_weeklyoff.fit_transform(df_clean["Weekly Off"])
df_clean["Significance_Encoded"] = le_significance.fit_transform(df_clean["Significance"])
df_clean["DSLR_Encoded"] = le_dslr.fit_transform(df_clean["DSLR Allowed"])
df_clean["BestTime_Encoded"] = le_besttime.fit_transform(df_clean["Best Time to visit"])
# Convert numerical columns
numeric_columns = [
    "Establishment Year",
    "time needed to visit in hrs",
    "Google review rating",
    "Entrance Fee in INR",
    "Number of google review in lakhs"
]
for col in numeric_columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
df_clean.dropna(inplace=True)
# Features for clustering
features = [
    "Establishment Year",
    "time needed to visit in hrs",
    "Google review rating",
    "Entrance Fee in INR",
    "Number of google review in lakhs",
    "Type_Encoded",
    "Significance_Encoded",
    "BestTime_Encoded"
]
X = df_clean[features]
# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Features scaled successfully.")
# ==========================================
# STEP 4: DATA MINING (K-Means)
# ==========================================
print("\n--- Step 4: K-Means Clustering ---")
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)
df_clean["Cluster"] = kmeans.fit_predict(X_scaled)
print("Clusters generated successfully.")
# ==========================================
# STEP 5: OUTPUT
# ==========================================
output_filename = "tourism_kmeans_output.csv"
df_clean.to_csv(output_filename, index=False)
print("\nOutput saved as:", output_filename)
print(df_clean[[
    "Name",
    "State",
    "City",
    "Type",
    "Google review rating",
    "Entrance Fee in INR",
    "Cluster"
]])
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# ==========================
# 1. SAMPLE
# ==========================

df = pd.read_csv("Top Indian Places to Visit.csv")

print("----- SAMPLE -----")
print("Total Records:", len(df))

# ==========================
# 2. EXPLORE
# ==========================

print("\n----- EXPLORE -----")
print(df.info())
print(df.describe(include="all"))

# ==========================
# 3. MODIFY
# ==========================

df = df.dropna().copy()

cat = [
    "Zone",
    "State",
    "City",
    "Type",
    "Airport with 50km Radius",
    "Weekly Off",
    "Significance",
    "DSLR Allowed",
    "Best Time to visit"
]

for c in cat:
    le = LabelEncoder()
    df[c] = le.fit_transform(df[c])

num = [
    "Establishment Year",
    "time needed to visit in hrs",
    "Google review rating",
    "Entrance Fee in INR",
    "Number of google review in lakhs"
]

for c in num:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df = df.dropna()

X = df[
    [
        "Establishment Year",
        "time needed to visit in hrs",
        "Google review rating",
        "Entrance Fee in INR",
        "Number of google review in lakhs",
        "Type",
        "Significance"
    ]
]

scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Data Modified Successfully")

# ==========================
# 4. MODEL
# ==========================

model = KMeans(n_clusters=3, random_state=42, n_init=10)

df["Cluster"] = model.fit_predict(X)

print("K-Means Model Created")

# ==========================
# 5. ASSESS
# ==========================

print(df[[
    "Name",
    "State",
    "City",
    "Google review rating",
    "Entrance Fee in INR",
    "Cluster"
]])

score = silhouette_score(X, df["Cluster"])
print("Silhouette Score:", score)

df.to_csv("tourism_semma_output.csv", index=False)

print("Output Saved Successfully")
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

print("--- PHASE 1: Business Understanding ---")
print("Objective: Segment tourist places based on features and visitor ratings.")


print("\n--- PHASE 2: Data Understanding ---")

df = pd.read_csv("Top Indian Places to Visit.csv")

print("Dataset Shape:", df.shape)
print(df.head())
df.info()


print("\n--- PHASE 3: Data Preparation ---")

df_prepared = df.dropna().copy()

cat_cols = [
    "Zone",
    "State",
    "City",
    "Type",
    "Airport with 50km Radius",
    "Weekly Off",
    "Significance",
    "DSLR Allowed",
    "Best Time to visit"
]

for col in cat_cols:
    le = LabelEncoder()
    df_prepared[col] = le.fit_transform(df_prepared[col])


num_cols = [
    "Establishment Year",
    "time needed to visit in hrs",
    "Google review rating",
    "Entrance Fee in INR",
    "Number of google review in lakhs"
]

for col in num_cols:
    df_prepared[col] = pd.to_numeric(df_prepared[col], errors="coerce")

df_prepared.dropna(inplace=True)


features = [
    "Establishment Year",
    "time needed to visit in hrs",
    "Google review rating",
    "Entrance Fee in INR",
    "Number of google review in lakhs",
    "Type",
    "Significance"
]

X = df_prepared[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Data Preparation Completed")


print("\n--- PHASE 4: Modeling ---")

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_prepared["Cluster"] = kmeans.fit_predict(X_scaled)

print("K-Means Clustering Completed")


print("\n--- PHASE 5: Evaluation ---")

score = silhouette_score(
    X_scaled,
    df_prepared["Cluster"]
)

print("Silhouette Score:", score)


print("\nCluster Profile:")

print(
    df_prepared.groupby("Cluster")[
        [
            "Establishment Year",
            "time needed to visit in hrs",
            "Google review rating",
            "Entrance Fee in INR"
        ]
    ].mean()
)


print("\n--- PHASE 6: Deployment ---")

output_file = "tourism_crisp_dm_output.csv"

df_prepared.to_csv(output_file, index=False)

print("Output saved:", output_file)
```
